**Exercise 6.**  PyTorch stores tensors in memory in row-major order ([read about it on Wikipedia](https://en.wikipedia.org/wiki/Row-_and_column-major_order)).  Essentially this means that the last index is incremented first, so the matrix so a 2D tensor, x, with shape $3 \times 2$ gets stored as x[0,0], x[0,1], x[1,0], x[1,1], x[2,0], x[2,1].  Notice that the last index changes first, then the next-to-last index, etc.  

`torch.reshape` iterates over the tensor elements in memory and fills them into the new shape in row-major order.
